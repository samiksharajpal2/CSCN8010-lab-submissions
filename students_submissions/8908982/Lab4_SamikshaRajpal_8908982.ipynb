## Practical Lab 4 - Multivariate Linear and Polynomial Regression, and Evaluation using R-Squared, MAPE and MAE.



### In this lab, we will run multivariate linear model and two polynomial regression models, and evaluate them using R-squared, Mean Absolute Percentage Error (MAPE) and Mean Absolute Error (MAE) metrices. The dataset will be Scikit-Learn's Diabetes dataset - the same one used in Practical Lab 3.

### Part 1: Get the data, and run a train-validation-test split. Description of each column can be found in sklearn documentation. Look at the documentation for the load_diabetes method to know what are as_frame and scaled arguments are for.

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plot
import seaborn as sb

import sklearn.datasets as datasets
from sklearn import metrics
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split 
from sklearn.metrics import mean_absolute_error, r2_score, mean_absolute_percentage_error

from sklearn.preprocessing import PolynomialFeatures
```

```python
X_diab, y_diab = datasets.load_diabetes(as_frame = True, scaled = False, return_X_y = True)
display(X_diab)
```

```python
display(y_diab)
```

### Part 2: Run a multivariate linear regression on all variables (1 point)

```python
X_train, X_test, y_train, y_test = train_test_split(X_diab, y_diab, test_size = 0.3, random_state = 42)
model_linear = LinearRegression()
model_linear.fit(X_train, y_train)

print("Intercept for Linear Regression : ",model_linear.intercept_)
print("Coffecients for Linear Regression : ",model_linear.coef_)

prediction = model_linear.predict(X_test)
```

### Polynomial regression 
### Part 3: Run a polynomial regression of the 2nd degree on the BMI feature alone (0.5 point)

```python
trainpoly = PolynomialFeatures(degree = 2, include_bias = False)
trainPolyLarge = trainpoly.fit_transform(X_train)
testPoly = PolynomialFeatures(degree = 2, include_bias = False)
testLargePoly = testPoly.fit_transform(X_test)
trainLargePoly = pd.DataFrame(trainPolyLarge, columns = trainpoly.get_feature_names_out())
display("Poly Train Large :::",trainPolyLarge)
testLargePoly = pd.DataFrame(testLargePoly, columns = testPoly.get_feature_names_out())
display("Poly Test Large ",testLargePoly)

XTrain_bmi_poly= trainLargePoly[['bmi','bmi^2']]
XTest_bmi_poly= testLargePoly[['bmi','bmi^2']]
```

```python
bmiModelPoly = LinearRegression()
bmiModelPoly.fit(XTrain_bmi_poly,y_train)
print("\n BMI degree 2 Intercept ::::",bmiModelPoly.intercept_)
print("\n BMI degree 2 - Coefficient for BMI & BMI^2 ::::",bmiModelPoly.coef_)
```

### Part 4. Run a polynomial regression of the 2nd degree on all features
This case is with 2 features considered, BMI and BMI^2 and a intercept. 

```python
fullModelPoly = LinearRegression()
fullModelPoly.fit(trainLargePoly, y_train)
print("\n All Features Degree 2 Intercept ::::",fullModelPoly.intercept_)
print("\n All Features Degree 2 Coefficient::::",fullModelPoly.coef_)
```

### Part 5. Compare the three models by looking at R-squared, MAPE and MAE. Explain what the values mean for a non-expert and add your insight about the values of each model. Note: You can add any further comparisons and code (this is not necessary for a perfect score, but will be reviewed and evaluated)

```python
print("Degree 1 - All Features")  
TrainDeg1All = model_linear.predict(X_train)          
TestDeg1All = model_linear.predict(X_test)
# Train
print("RSqrTrainDegree1",r2_score(y_train, TrainDeg1All))
print("MAETrainDegree1",mean_absolute_error(y_train, TrainDeg1All))
print("MAPTrainDegree1", mean_absolute_percentage_error(y_train, TrainDeg1All))
# Test
print("R2TestDegree1", r2_score(y_test, TestDeg1All))
print("MAETestDegree1",  mean_absolute_error(y_test, TestDeg1All))
print("MAPTestDegree1",mean_absolute_percentage_error(y_test, TestDeg1All),"\n")

print("BMI feature alone- Degree 2")  
BMIpredictionTrainDegree2 = bmiModelPoly.predict(XTrain_bmi_poly) 
BMIpredictionTestDegree2 = bmiModelPoly.predict(XTest_bmi_poly)
# Train 
print("BMIR2TrainDegree2 ",r2_score(y_train, BMIpredictionTrainDegree2))
print("BMIMAETrainDegree2", mean_absolute_error(y_train, BMIpredictionTrainDegree2))
print("BMIMAPTrainDegree2", mean_absolute_percentage_error(y_train, BMIpredictionTrainDegree2))
# Test
print("BMIR2TestDegree2 = ",r2_score(y_test, BMIpredictionTestDegree2))
print("BMIMAETestDegree2 =", mean_absolute_error(y_test, BMIpredictionTestDegree2))
print("BMIMAPTestDegree2 =", mean_absolute_percentage_error(y_test, BMIpredictionTestDegree2),"\n")

print("All Features - Degree 2")  
TrainDeg2All = fullModelPoly.predict(trainLargePoly)  
TestDeg2All = fullModelPoly.predict(testLargePoly)
# Train 
print("R2TrainDegree2 = ",r2_score(y_train, TrainDeg2All))
print("MAETrainDegree2 =", mean_absolute_error(y_train, TrainDeg2All))
print("MAPTrainDegree2 =", mean_absolute_percentage_error(y_train, TrainDeg2All))
# Test
print("RSqrTestDegree2 = ",r2_score(y_test, TestDeg2All))
print("MAETestDegree2 =", mean_absolute_error(y_test, TestDeg2All))
print("MAPTestDegree2 =", mean_absolute_percentage_error(y_test, TestDeg2All),"\n")

```
